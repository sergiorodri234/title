```python
for col, por in ((df_filter.isna().sum() / df_filter.shape[0]) * 100).to_dict().items():
  print(f"En la columna '{col}' hace falta el {round(por,4)}% de los datos")
```

```python
import numpy as np
import scipy.stats as st
from scipy.spatial.distance import mahalanobis
mahal_distances = []

for row in df_filter.to_numpy():
  mahal_distances.append(mahalanobis(row, df_filter.mean(), np.linalg.inv(np.cov(df_filter.values.T))))
```

```python
k = df_filter.shape[0]

df_filter['mahal_distances'] = mahal_distances
df_filter['p_value'] = 1 - st.chi2.cdf(df_filter['mahal_distances'], k-1)
```

```python
df_filter = df_filter[df_filter['p_value'] > 0.05]
```

```python
df_filter.drop(columns=['mahal_distances','p_value'],inplace=True)
```
